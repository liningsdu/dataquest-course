## How far off from the mean a value of 220000 is in the distribution of the SalePrice variable.

```python
houses.SalePrice.plot.kde(xlim = (houses.SalePrice.min(), houses.SalePrice.max()))
plt.axvline(houses.SalePrice.mean(), color = 'Black', label = 'Mean')
plt.axvline(houses.SalePrice.mean() + houses.SalePrice.std(ddof = 0), color = 'Red', label = 'Standard deviation')
plt.axvline(220000, color = 'Orange', label = '220000')
plt.legend()
plt.show()
```

```python
st_devs_away = (220000 - houses.SalePrice.mean()) / houses.SalePrice.std(ddof=0)
st_devs_away
```

* For any value x in a populaton with mean $\mu$ and standard deviation $\sigma$, the number of standard deviations wary from the mean:
\begin{equation}
z = \frac{x - \mu}{\sigma}
\end{equation}
***
## The value representing the number of standard deviations away from the mean is commonly known as the standard score, or the **z-score**.
* For samples:
\begin{equation}
z = \frac{x - \bar{x}}{s}
\end{equation}

## A z-score has two parts:
   * The sign, which indicates whether the value is above or below the mean
   * The value, which indicates the number of standard deviations that a value is away from the mean.

```python
def get_zscore(value, array, degree = 0):
    # ddof 0 for population, and 1 for sample
    mean = sum(array) / len(array)
    distance = value - mean
    std = np.std(array, ddof = degree)
    z_score = distance / std
    return z_score
min_val = houses.SalePrice.min()
mean_val = houses.SalePrice.mean()
max_val = houses.SalePrice.max()

min_z = get_zscore(min_val, houses.SalePrice)
mean_z = get_zscore(mean_val, houses.SalePrice)
max_z = get_zscore(max_val, houses.SalePrice)

print(min_z, mean_z, max_z)
```

```python
def z_score(value, array, bessel= 0):
    return get_zscore(value, array, bessel)
```

```python
z_anmes = z_score(200000, houses.loc[houses.Neighborhood == 'NAmes', 'SalePrice'])
z_collgcr = z_score(200000, houses.loc[houses.Neighborhood == 'CollgCr', 'SalePrice'])
z_oldtown = z_score(200000, houses.loc[houses.Neighborhood == 'OldTown', 'SalePrice'])
z_edwards = z_score(200000, houses.loc[houses.Neighborhood == 'Edwards', 'SalePrice'])
z_somerst = z_score(200000, houses.loc[houses.Neighborhood == 'Somerst', 'SalePrice'])

print(z_anmes, z_collgcr, z_oldtown, z_edwards, z_somerst)
```

```python
houses.SalePrice.plot.kde(xlim = (houses.SalePrice.min(), houses.SalePrice.max()))
```

```python
mean  = houses.SalePrice.mean()
st_dev = houses.SalePrice.std(ddof=0)
houses['z_prices'] = houses.SalePrice.apply(lambda x: (x-mean)/st_dev)

plt.figure(figsize=(11,3.5))
plt.subplot(1,2,1)
houses['z_prices'].plot.kde(xlim = (houses.z_prices.min(), houses.z_prices.max()))
plt.subplot(1,2,2)
houses.SalePrice.plot.kde(xlim = (houses.SalePrice.min(), houses.SalePrice.max()))
plt.tight_layout() #otherwise the plots will overlay paritally
```

```python
z_mean_price = houses.z_prices.mean()
z_mean_price
```

```python
z_stdev_price = houses.z_prices.std(ddof = 0)
z_stdev_price
```

```python
mean_area = houses['Lot Area'].mean()
st_dev_area = houses['Lot Area'].std(ddof = 0)
houses['z_lot_area'] = houses['Lot Area'].apply(lambda x : (x - mean_area) / st_dev_area)
```

```python
z_mean_area = houses.z_lot_area.mean()
z_stdev_area = houses.z_lot_area.std(ddof = 0)
print(z_mean_area, z_stdev_area)
```

```python
houses.z_lot_area.plot.kde(xlim = (houses.z_lot_area.min(), houses.z_lot_area.max()))
```

## Both the mean values extremely close to 0. For SalePrice, mean of z-score is -1.14e-16, Lot area is 8.77e-18, also, both the standard deviations were also very close to 1.
***
## For every distribution of z-scores, the mean is always 0 and the standard deviation is always 1.
## A distribution of z-score is often called a standard distribution. When we convert a distribution to z-scores, we'd say in statistical jargon that we standardized the distribution.

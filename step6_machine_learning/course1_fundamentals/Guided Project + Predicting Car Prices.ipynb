# Predicting Car Prices using k-nearest neighbors algorithm

## The goal for this project is to practice the machine learning workflow you've learned so far to predict a car's market price using its attributes.
Data source: https://archive.ics.uci.edu/ml/datasets/automobile, can also be directly downloaded from: https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data

```python
import numpy as np
import pandas as pd
pd.options.display.max_columns = 100
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
column_names = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 
               'body-style', 'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height',
               'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore',
               'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars = pd.read_csv('imports-85.data', header=None, names= column_names)
```

## Columns:  engine-size, compression-ratio, horsepower, city-mpg, highway-mpg are numeric and can be used as features, column price is the target column

```python
# replace ? with numpy.nan
cars.replace('?', np.nan, inplace=True)
```

```python
# Convert to numeric columns
columns_to_convert = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
cars[columns_to_convert] = cars[columns_to_convert].astype('float')
cars.info()
```

## Remove non-numeric columns and non-cotinuous columns

```python
columns_numeric = cars.columns[(cars.dtypes == 'int64') | (cars.dtypes == 'float64')]
columns_numeric
```

```python
cars_numeric = cars[columns_numeric].copy()
cars_numeric.drop('symboling', axis = 1, inplace = True)
cars_numeric.isnull().sum()
```

## Remove rows conaining null value

```python
cars_numeric = cars_numeric.astype('float')
cars_numeric.dropna(subset = ['price'], inplace=True)
cars_numeric.info()
```

```python
# Fill the other null values with the mean value of the column
cars_numeric = cars_numeric.fillna(cars_numeric.mean())
```

## Normalize the numeric values of feature columns, with the mean of 0 and standard deviation of 1
* To prevent any single column from having too much of an impact on the distance, we can **normalize** all of the columns to have a mean of 0 and a standard deviation of 1.
   * from each value, substract the mean of the column
   * divide each value by the standard deviation of the column.
   \begin{equation}
   x = \frac{x-\mu}{\sigma}
   \end{equation}
   x is a value in a specific column, $\mu$ is the mean of all the values in the column, $\sigma$ is the standard deviation ofall the value in the column

```python
columns_to_normalize = cars_numeric.columns[cars_numeric.columns != 'price']
cars_normalized = cars_numeric[columns_to_normalize].copy()
cars_normalized = (cars_normalized - cars_normalized.mean()) / cars_normalized.std()
cars_normalized['price'] = cars_numeric['price']
cars_normalized.head()
```

## Normalize the feature numeric ones so all values range from 0 to 1

```python
cars_normalized_0_1 = cars_numeric[columns_to_normalize].copy()
cars_normalized_0_1 = (cars_normalized_0_1 - cars_normalized_0_1.min()) / (cars_normalized_0_1.max() - 
                                                                          cars_normalized_0_1.min())
cars_normalized_0_1['price'] = cars_numeric['price']
cars_normalized_0_1.head()
```

```python
cars_normalized_0_1.describe()
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Function that encapsulates the training and simple validation process.
def knn_train_test(training_columns, target_column, df):
    # training_columns should be a list
    # split the dataset
    #random_index = np.random.permutation(df.shape[0])
    df_length = df.shape[0]
    cutoff = int(df_length * 0.75)
    train_df = df.iloc[0:cutoff]
    test_df = df.iloc[cutoff:]
    
    # Instantiate the KNeighborsRegressor class
    model_knn = KNeighborsRegressor()
    model_knn.fit(train_df[training_columns], train_df[target_column])
    predictions = model_knn.predict(test_df[training_columns])
    
    # Calculate and return the RMSE
    mse = mean_squared_error(test_df[target_column], predictions)
    rmse = np.sqrt(mse)
    return rmse
```

## Train and test univariate model using the different numeric columns in the data set

```python
training_columns = cars_normalized_0_1.columns[cars_normalized_0_1.columns != 'price']
target_column = 'price'
random_index = np.random.permutation(cars_normalized_0_1.index)
cars_df = cars_normalized_0_1.loc[random_index].copy()
# Use cars_df for prediction, since in original data, cars data are clustered by make column
```

```python
rmses = []

for column in training_columns:
    rmse = knn_train_test([column], target_column, cars_df)
    rmses.append((column, rmse))
    print('Using column {} to predict price, RMSE value is {}.'.format(column, rmse))
sorted(rmses, key=lambda x:x[-1])
```

* Engine size performed the best using the default k value.

## Modify the knn_train_test() to accept a parameter for the k value

```python
def knn_train_test(training_columns, target_column, df, k = 5):
    # training_columns should be a list
    # split the dataset
    #random_index = np.random.permutation(df.shape[0])
    if isinstance(training_columns, str):
        training_columns = [training_columns]
    df_length = df.shape[0]
    cutoff = int(df_length * 0.75)
    train_df = df.iloc[0:cutoff]
    test_df = df.iloc[cutoff:]
    
    # Instantiate the KNeighborsRegressor class
    model_knn = KNeighborsRegressor(n_neighbors=k)
    model_knn.fit(train_df[training_columns], train_df[target_column])
    predictions = model_knn.predict(test_df[training_columns])
    
    # Calculate and return the RMSE
    mse = mean_squared_error(test_df[target_column], predictions)
    rmse = np.sqrt(mse)
    return rmse
```

```python
k_train_df = pd.DataFrame({'k_value' : [1,3,5,7,9]})
for name in training_columns:
    k_train_df[name] = 0
k_train_df
```

```python
for k in range(1,11, 2):
    for column_name in training_columns:
        k_train_df.loc[k_train_df['k_value'] == k, column_name] = knn_train_test(column_name, target_column, 
                                                                                cars_df, k)
k_train_df
```

```python
plt.figure(figsize=(10,8))
for name in training_columns:
    plt.plot(k_train_df['k_value'], k_train_df[name], label = name)
    
plt.ylim(0,9500)
plt.legend(loc = 'best')
plt.xlabel('k_value', fontsize = 15)
plt.ylabel('RMSE (dollars)', fontsize = 15)
plt.show()
```

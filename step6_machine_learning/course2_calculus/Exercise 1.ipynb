```python
def slope(x_1, x_2, y_1, y_2):
    return (y_2 - y_1) / (x_2 - x_1)
```

* A linear that intersects 2 points on a curve is known as a **secant line**

```python
import seaborn as sns
sns.set(style='darkgrid')
```

```python
def draw_secant(x_values):
    x = np.linspace(-20,30,100)
    y = -1 * (x ** 2) + x * 3 - 1
    plt.plot(x, y)
    
    x_1 = x_values[0]
    x_2 = x_values[1]
    y_1 = -1* (x_1 ** 2) + x_1 * 3 -1
    y_2 = -1 * (x_2 ** 2) + x_2 * 3 - 1
    m = (y_2 - y_1) / (x_2 - x_1)
    b = y_1 - m * x_1
    y = m * x + b
    plt.plot(x, y, color = 'g')
    
    plt.show()
    
```

```python
draw_secant([3,15])
```

## Derivative
* A derivative is the slope of the tangent line at any point along a curve.
* A function's derivative can tell us the slope of the tangent line for any x value along the function.
* The process of finding a function's derivative is known as **differentiation**.

```python
x = np.linspace(-5, 6, 110)
y = -1 * 2 * x + 3
plt.plot(x, y)
```

## Critical point
* At a function's critical point, the slope is either 0 or undefined.
* **critical points vs. extreme values**
*  critical point can be a minimum value if the slope transitions from negative to positive at that point. When the slope transitions from positive to negative at a point, it can be a maximum value. If the slope doesn't transition at a critical point, like at for in the last screen, then it can't be a minimum or a maximum point.
* A curve can have multiple critical points and multiple candidates for minimum and maximum values.
* If a critical point is the lowest point in the interval surrounding it, it's known as a **relative minimum**. If it's the highest point in the neighborhood, it's known as a **relative maximum**.
